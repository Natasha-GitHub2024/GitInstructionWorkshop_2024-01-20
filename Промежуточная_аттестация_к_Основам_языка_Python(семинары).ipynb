**1.2. Изучите стоимости недвижимости.
Для этого можно построить график зависимости цены дома от площади**

```python
plt.figure(figsize=(8,6))
plt.scatter(x=kc_houses['sqft_lot'], y=kc_houses['price'])
plt.title("Зависимость цены дома от площади")
plt.xlabel("Площадь (кв. метры)")
plt.ylabel("Цена")
plt.show()
```

**Вывод:** Стоимость жилья растёт с увеличением площади

**1.3. Изучите распределение квадратуры жилой площади.
Можно посмотреть на гистограмму распределения квадратных метров.**

```python
plt.figure(figsize=(8,6))
plt.hist(kc_houses['sqft_living'], bins=20)
plt.title("Распределение площади жилых помещений")
plt.xlabel("Площадь (кв. метры)")
plt.ylabel("Количество")
plt.show()
```

**Вывод:** Большинство квартир имеют площадь около 1900 кв. метров.

**1.4. Изучите распределение года постройки**

```python
plt.figure(figsize=(8,6))
plt.hist(kc_houses['yr_built'], bins=20)
plt.title("Распределение годов постройки")
plt.xlabel("Год постройки")
plt.ylabel("Количество объектов")
plt.show()
```

**Вывод:** Наибольшее количество объектов построено в период с 50-х до 90-х годов.

**Задача 2**

**2.1. Изучение распределения домов от наличия вида на набережную.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.boxplot (x = kc_houses['waterfront'], y = kc_houses['price'], showfliers = False)
plt.title("Распределение цен по виду на набережную")
plt.xlabel("Вид с объекта недвижимости")
plt.ylabel("Цена")
plt.xticks([0,1], ['Вид без набережной', 'Вид на набережную'])
plt.show()
```

**Вывод:** Дома с видом на набережную стоят дороже

**2.2. Изучение распределения этажей домов**

```python
plt.figure(figsize=(8,6))
plt.hist(kc_houses['floors'], bins=20)
plt.title("Распределение количества этажей")
plt.xlabel("Этажи")
plt.ylabel("Количество объектов")
plt.grid()
plt.show()
```

```python
kc_houses['floors'].value_counts()
```

**Вывод:** Преобладание одноэтажных и двухэтажных домов

**2.3. Изучение состояния домов**

```python
plt.figure(figsize=(8,6))
plt.hist(kc_houses['condition'], bins=20)
plt.title("Состояние недвижимости")
plt.xlabel("Состояние")
plt.ylabel("Количество")
plt.grid()
plt.show()
```
